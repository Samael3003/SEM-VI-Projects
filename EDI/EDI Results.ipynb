# Solar Flares...

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('solar_flare_readings.csv', parse_dates=['Timestamp'])

# Plot the voltage readings
plt.figure(figsize=(10, 6))
plt.plot(df['Timestamp'], df['Voltage (mV)'], color='blue', marker='o', linestyle='-')
plt.title('Voltage Readings during Solar Flare Observation')
plt.xlabel('Timestamp')
plt.ylabel('Voltage (mV)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('solar_intensity_vs_temperature.csv')

# Plotting the data
plt.figure(figsize=(20, 6))
plt.plot(df['Time'], df['Solar Radiation Intensity (dB)'], label='Solar Radiation Intensity (dB)', color='orange')
plt.plot(df['Time'], df['Ambient Temperature (Celsius)'], label='Ambient Temperature (Celsius)', color='blue')

# Adding labels and title
plt.xlabel('Time (minutes)')
plt.ylabel('Intensity / Temperature')
plt.title('Solar Radiation Intensity vs Ambient Temperature')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

# Intelsat-12 (241, 51)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('azimuth_elevation_signal_strength1.csv', index_col=0)

# Plotting the heatmap
plt.figure(figsize=(10, 6))
plt.imshow(df, cmap='magma', aspect='auto', extent=[df.columns[0], df.columns[-1], df.index[-1], df.index[0]])

# Adding colorbar
plt.colorbar(label='Signal Strength')

# Adding labels and title
plt.xlabel('Azimuth')
plt.ylabel('Elevation')
plt.title('Frequency Map of Satellite Signal Strength')

# Display the plot
plt.show()
```
